```
import csv
import pandas as pd
df = pd.read_csv('movies_1980_2020_30k.csv')
director_counts = df.groupby('Director').count()
unique_directors = df['Director'].nunique()
print(f"Total unique directors: {unique_directors}")
top_directors = df['Director'].value_counts().head(10)
print(top_directors)
best_rated_movies = df.groupby('Title')['Rating'].mean().sort_values(ascending=False)
print(best_rated_movies.head(10))
best_rated_movies = df.groupby('Title').agg({'Rating': 'mean', 'Director': 'first'})
best_rated_movies = best_rated_movies.sort_values(by='Rating', ascending=False)
print(best_rated_movies.head(10))

```

```
import csv
import pandas as pd
df = pd.read_csv('movies_1980_2020_30k.csv')
filtered_df = df[df['Rating'] > 7]
genre_counts = filtered_df.groupby('Genre').count()
genre_counts = genre_counts.sort_values(by='Rating', ascending=False)
print(genre_counts)
import seaborn as sns
import matplotlib.pyplot as plt
seabornchart = sns.barplot(x=genre_counts.index, y=genre_counts['Rating'])
plt.show()

```

```
# Create a count plot
plt.figure(figsize=(10, 6))
#sns.countplot(x='Rating', data=df, order=df['Rating'].value_counts().nlargest(10).index)
sns.countplot(x='Genre', data=df, order=df['Genre'].value_counts().index)


# Customize the plot
plt.title('Most Common Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
# Get the top 10 most common ratings
```

```
genre_ratings = df.groupby('Genre')['Rating'].mean().sort_values(ascending=False)

# Display the results
print(genre_ratings)

# Get the genre with the highest average rating
top_genre = genre_ratings.index[0]
top_rating = genre_ratings.iloc[0]

# Group by Genre and calculate mean Rating
genre_ratings = df.groupby('Genre')['Rating'].mean().sort_values(ascending=True)

# Create a figure and axes
plt.figure(figsize=(12, 8))

# Create a horizontal bar plot
sns.barplot(x=genre_ratings.values, y=genre_ratings.index, orient='h')

# Customize the plot
plt.title('Average Ratings by Genre', fontsize=16)
plt.xlabel('Average Rating', fontsize=12)
plt.ylabel('Genre', fontsize=12)

# Add value labels to the end of each bar
for i, v in enumerate(genre_ratings.values):
  plt.text(v, i, f'{v:.2f}', va='center')

# Adjust layout to prevent cutting off labels
plt.tight_layout()

# Show the plot
plt.show()
```
